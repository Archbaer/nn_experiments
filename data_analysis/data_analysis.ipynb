```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("countries.csv")
df
```

```python
top_countries = df.sort_values(by="Population (millions)", ascending=False).head(10)
top_countries
```

```python
plt.xlabel("Country")
plt.ylabel("Population (millions)")
plt.bar(top_countries["Country"], top_countries["Population (millions)"])
plt.xticks(rotation=90)
```
