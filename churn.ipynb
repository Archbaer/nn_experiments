```python
import pandas as pd

def create_temporal_features(train):
    temporal_rows = []

    for _, row in train.iterrows():
        customer_id = row['Customer ID']
        tenure = row['Tenure in Months']
        churn = row['Churn']
        # For each month up to tenure, create a snapshot
        for t in range(1, tenure):
            snapshot = row.copy()
            snapshot['temporal_index'] = t
            # Label: did they churn at t+1? (for real data, you need to know this)
            snapshot['churn_next_month'] = 1 if (t + 1 == tenure and churn == 1) else 0
            temporal_rows.append(snapshot)

    temporal_df = pd.DataFrame(temporal_rows)
    return temporal_df
```

```python
temporal_train = create_temporal_features(train)
temporal_test = create_temporal_features(test)
temporal_validation = create_temporal_features(validation)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
import numpy as np

# Create a combined preprocessor function
def preprocess_all_datasets(X_train, X_validation, X_test, y_train, y_validation, y_test):
    # Identify column types from training data

    cat_col = X_train.select_dtypes(include=['object']).columns.tolist()
    num_col = X_train.select_dtypes(include=['int64', 'float64']).columns.tolist()
    
    # Create and fit preprocessor on training data only
    preprocessor = ColumnTransformer(
        transformers=[
            ('num', StandardScaler(), num_col),
            ('cat', OneHotEncoder(handle_unknown='ignore'), cat_col)
        ]
    )
    
    # Fit on training data
    preprocessor.fit(X_train)
    
    # Transform all datasets with the same preprocessor
    X_train_processed = preprocessor.transform(X_train)
    X_validation_processed = preprocessor.transform(X_validation)
    X_test_processed = preprocessor.transform(X_test)
    
    # Convert to dense arrays if needed
    if not isinstance(X_train_processed, (np.ndarray, np.generic)):
        X_train_processed = X_train_processed.toarray()
        X_validation_processed = X_validation_processed.toarray()
        X_test_processed = X_test_processed.toarray()
    
    # Convert targets to numpy arrays
    y_train_processed = np.array(y_train)
    y_validation_processed = np.array(y_validation)
    y_test_processed = np.array(y_test)
    
    return (X_train_processed, X_validation_processed, X_test_processed, 
            y_train_processed, y_validation_processed, y_test_processed)
```

```python
X_train = temporal_train.drop(['Churn', 'churn_next_month'], axis=1)
y_train = temporal_train['churn_next_month']

X_test = temporal_test.drop(['Churn', 'churn_next_month'], axis=1)
y_test = temporal_test['churn_next_month']

X_validation = temporal_validation.drop(['Churn', 'churn_next_month'], axis=1)
y_validation = temporal_validation['churn_next_month']
```

```python
X_train, X_validation, X_test, y_train, y_validation, y_test = preprocess_all_datasets(
    X_train, X_validation, X_test, y_train, y_validation, y_test
)
```

```python
from sklearn.utils import shuffle
def shuffler(X, y):
    X, y = shuffle(X, y, random_state=42)
    return X, y

X_train, y_train = shuffler(X_train, y_train)
X_test, y_test = shuffler(X_test, y_test)
X_validation, y_validation = shuffler(X_validation, y_validation)
```

```python
from xgboost import XGBClassifier


classifier = XGBClassifier(
    objective='binary:logistic',
    eval_metric=['logloss', 'auc', 'error'],
    tree_method='hist',
    device='cuda'
)
```

```python
classifier.fit(
    X_train, y_train,
    eval_set=[(X_train, y_train), (X_validation, y_validation)],
    verbose=True
)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc, precision_recall_curve, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import seaborn as sns

y_test_pred = classifier.predict(X_test)

print(classification_report(y_test, y_test_pred))
print(confusion_matrix(y_test, y_test_pred))

ConfusionMatrixDisplay.from_estimator(classifier, X_test, y_test)
```

```python
cm = confusion_matrix(y_train[:43668], y_test_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['No Churn', 'Churn'],
            yticklabels=['No Churn', 'Churn'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix (Training Set)')
plt.show()
```

```python
val_pred = classifier.predict(X_validation)
print(classification_report(y_validation, val_pred))
```

```python
fpr, tpr, _ = roc_curve(y_train[:43668], y_test_pred)
train_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'Training (AUC = {train_auc:.4f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```
